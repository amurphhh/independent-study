## Let's try the Mann Whitney U test

```python
import scipy
from scipy import stats
```

```python
#scipy.stats.mannwhitneyu(x,y, alternative = )

# Comparing Comfort and Relaxation during the ride
print(list(data2)[2])
x = data2.loc[data2['Intervention']=='Julie'][list(data2)[2]]
y = data2.loc[data2['Intervention']=='Lily'][list(data2)[2]]
scipy.stats.mannwhitneyu(x,y, alternative = 'greater')
```

```python
# Comparing UX_Score
print(list(data2)[57])
x = data2.loc[data2['Intervention']=='Julie'][list(data2)[57]]
y = data2.loc[data2['Intervention']=='Lily'][list(data2)[57]]
scipy.stats.mannwhitneyu(x,y, alternative = 'greater')
```

```python
# Comparing Confidence Score
print(list(data2)[58])
x = data2.loc[data2['Intervention']=='Julie'][list(data2)[58]]
y = data2.loc[data2['Intervention']=='Lily'][list(data2)[58]]
scipy.stats.mannwhitneyu(x,y, alternative = 'greater')
```

Takeaways:
    
    Individual columns likely don't show a statistically significant difference (example - column 2 for Comfort & relaxation during the ride). 
    
    However, certain scores that are combined, such as the Confidence Score, calculated as the sum of the following columns, is significant.

###  Let's try this

Can I run the MannWhittneyU test on each column via a for loop? and print out each result. 

```python
for i in range(1,len(list(data2))):
    if isinstance(data2[list(data2)[i]][26], (int, np.int64)):
        print(i)
        print(list(data2)[i])
        x = data2.loc[data2['Intervention']=='Julie'][list(data2)[i]]
        y = data2.loc[data2['Intervention']=='Lily'][list(data2)[i]]
        print(scipy.stats.mannwhitneyu(x,y, alternative = 'greater'))
    
```

### All Significant Tests

```python
for i in range(1,len(list(data2))):
    if isinstance(data2[list(data2)[i]][26], (int, np.int64)):
        x = data2.loc[data2['Intervention']=='Julie'][list(data2)[i]]
        y = data2.loc[data2['Intervention']=='Lily'][list(data2)[i]]
        if scipy.stats.mannwhitneyu(x,y, alternative = 'greater').pvalue < .05:
            print(i)
            print(list(data2)[i])
            print(scipy.stats.mannwhitneyu(x,y, alternative = 'greater'))
    
```

### Findings

Only 6 of the columns report a significant difference between Julie and Lily, with 1 of the aggregate column scores also being significant. 

The user's 'Confidence Score' - a sum of the user's score for questions rating their confidence in the vehicle's capacity in terms of AV Control, Driving Capability, and Decision-making ability is the most significant. 

How useful the information presented was, the digital assistant helped make the vehicle seem Trustworthy were next. 

Being asked how they felt throughout the ride and Future Preference for a digital assistant was also significant - the drivers prefer Julie.
